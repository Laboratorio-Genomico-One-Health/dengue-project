```r
library('imputeTS')
library('TSA')
daily <- read.csv(file = './local/data/medellin_consolidado_2012_2020.csv')
head(daily)
```

```r
help(harmonic)
```

```r
temp = daily['TEMPERATURE_MEAN']
temp <- ts(temp, frequency = 52, start = 2012)
time.pts = c(1:length(temp))
time.pts = c(time.pts - min(time.pts))/max(time.pts)
```

```r
x1 = time.pts
x2 = time.pts^2
lm.fit = lm(temp~x1+x2)
temp.fit.lm = ts(fitted(lm.fit),start=2012,frequency=52)
ts.plot(temp,ylab="Temperature")
lines(temp.fit.lm,lwd=2,col="green")
abline(temp.fit.lm[1],0,lwd=2,col="blue")
```

```r
temp <- ts(temp, frequency = 52, start = 2012)
month = season(temp)
```

```r
har1=harmonic(temp,1)
model3=lm(temp~har1)
summary(model3)
```

```r
har2=harmonic(temp,2)
model4=lm(temp~har2)
summary(model4)
```

```r
model1 = lm(temp~month)
model2 = lm(temp~month-1)
st1 = coef(model2)
```

```r
st2 = fitted(model4)[1:52]
```

```r
plot(1:52,st1,lwd=2,type="l",xlab="Month",ylab="Seasonality")
lines(1:52,st2,lwd=2, col="brown")
```
